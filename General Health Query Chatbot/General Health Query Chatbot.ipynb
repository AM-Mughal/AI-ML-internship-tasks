```python
SYSTEM_PROMPT = """
You are a helpful and cautious medical assistant.

Rules:
- Only provide general health information.
- Do NOT diagnose diseases.
- Do NOT prescribe medication or exact dosages.
- Always recommend consulting a doctor for serious concerns.
- Keep responses simple, clear, and friendly.

Tone:
- Supportive and easy to understand.
"""

DANGEROUS_KEYWORDS = [
    "dose", "dosage", "overdose",
    "suicide", "kill myself", "emergency", "severe pain",
    "chest pain", "unconscious"
]

def is_dangerous(query):
    return any(word in query.lower() for word in DANGEROUS_KEYWORDS)

def safe_fallback():
    return """
⚠️ I can't provide specific medical advice.

Please consult a doctor or a qualified healthcare professional.
If this is urgent, seek medical help immediately.
"""
```

## Model

```python
model = genai.GenerativeModel("gemini-2.5-flash")
```

```python
import time
def get_response(user_query):
    if is_dangerous(user_query):
        return safe_fallback()

    full_prompt = SYSTEM_PROMPT + "\nUser: " + user_query

    for i in range(3):
        try:
            response = model.generate_content(full_prompt)
            return response.text
        except Exception as e:
            print(f"Retry {i+1}... waiting")
            time.sleep(10)

    return "⚠️ API is busy or quota exceeded. Try again later."
```

```python
print("🩺 Health Chatbot (type 'exit' to stop)\n")

while True:
    query = input("You: ")

    if query.lower() == "exit":
        print("Bot: Take care! 👋")
        break

    answer = get_response(query)
    print("Bot:", answer)
```
